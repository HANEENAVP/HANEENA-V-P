# CASE STUDY ON TESTING OF HYPOTHESIS

```python
data1=pd.read_csv(r"C:\Users\hanee\Downloads\Sales_add.csv")
data1
```

```python
#here we use the t test
```

```python
from scipy.stats import ttest_ind
from scipy.stats import chi2_contingency
```

# a)

```python
#H0: Digital marketing has no effect on sales.
#H1: Digital marketing has effect on sales.
```

```python
t,pv = ttest_ind(data1['Sales_before_digital_add(in $)'],data1['Sales_After_digital_add(in $)'])
```

```python
if pv<0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null Hypothesis')
```

```python
so we can conclude that  digital marketing has effect on sales,thta is there is a increament
```

# b)

```python
data2 = pd.crosstab(data1['Region'],data1['Manager'])
```

```python
chi, p_v,_,_ = chi2_contingency(data2)
```

```python
if p_v<0.05:
    print("Rejecting Independence")
else:
    print("Accepting Independence")
```

so we can conclude that 'region' and 'manager' are independent of each other
